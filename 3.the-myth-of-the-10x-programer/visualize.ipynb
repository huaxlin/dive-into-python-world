```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame([
    ('(500,500.0,  601, 401)',   1.893,  0.244, 0.692798),
    ('(500,500.0, 5601,5401)', 239.478, 51.416, 84.9712),
], columns=['params', 'Pure-Python', 'Numpy-Python', 'C++ (std:11)'])
```

```python
df
```

```python
df.plot.bar(
    x="params",
    y=["Pure-Python", "Numpy-Python", "C++ (std:11)"],
    rot=0
)
plt.yscale("log")
plt.show()
```
